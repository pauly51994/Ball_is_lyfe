```python
import requests
from bs4 import BeautifulSoup
import time
from tqdm import tqdm
import pandas as pd
import pickle
```

# Scrape data

###### parse html of any url provided

```python
def getAndParseUrl(url):
    results = requests.get(url)
    time.sleep(1)
    soup = BeautifulSoup(results.text, 'html.parser')
    return soup
```

```python
def get_player_pos():
    
    import string
    
    # container for each player and his position
    store = []
    
    # load player ids for basketball-reference.com
    player_ids = pickle.load(open('player_ids.pickle','rb'))

    # loop through each letter in english alphabet
    for let in list(string.ascii_lowercase):
        
        # loop through each player
        for pid in tqdm([p for p in player_ids if p.startswith(let)]):
            soup = getAndParseUrl(f'https://www.basketball-reference.com/players/{let}/{pid}/gamelog/2019')
            
            player = {'p_id': pid,
                      'pos': [l for l in [t.text.split('\n') for t in soup.select('#meta p')] if '  Position:' in l][0][4].strip()}
            
            store.append(player)
            
            # pickle store each time a player is added so data will not be lost if error occurs
            with open('player_positions.pickle','wb') as outfile:
                pickle.dump(store, outfile)
                
            outfile.close()
            
        # check length of store to make sure player is getting added each time
        print(len(store))
        
    return store
```

```python
player_pos = get_player_pos()
```
