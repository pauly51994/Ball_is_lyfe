###### get player_id from basketball reference to plug into the url parameter for getting games

```python
def get_player_ids():
    
    import string
    
    # create a list of the english alphabet
    alpha = list(string.ascii_lowercase)
    
    # container for every possible players' id
    player_id = []
    
    # this url lists every player by first-letter of last name
    # each letter is a different page of players
    # so loop through alphabet to get every player id
    for letter in tqdm(alpha):
        url = f'https://www.basketball-reference.com/players/{letter}/'
        soup = getAndParseUrl(url)
        name_soup = soup.select('tr th.left')
        
        # extract each player id from the soup
        for name in name_soup:
            player_id.append(name.attrs['data-append-csv'])
            
    # return the list of player ids
    return player_id
```

```python
def get_games(year=2019):
    import string
    
    # list of english alphabet
    alpha = list(string.ascii_lowercase)
    
    # list container for each game for every player
    games = []
    
    # loop through alphabet
    for letter in alpha:
        
        # group player ids by first-letter of last name
        ids = [p for p in player_ids if p.startswith(letter)]
        
        # loop through each player id of grouped
        for player in tqdm(ids):
            # change url to access each player's game log for any year provided
            url = f'https://www.basketball-reference.com/players/{letter}/{player}/gamelog/{year}'

            # each url is a player's game log for the season
            player_soup = getAndParseUrl(url)

            # get each row from the url
            rows = player_soup.select('#pgl_basic tbody tr')
            
            # each row is a game
            for row in tqdm(rows):

                # container to hold all stats for a single game
                single = {}

                # make sure each single game has player id (good for when creating dataframe)
                single['player_id'] = player
                
                #make sure each single player game has player name (used to combine positions)
                single['player'] = player_soup.select('#meta h1')[0].text[:-17]

                # select the td tags to get all the stats
                soup = row.select('td')

                # go through each element of jbsoup (each el is a single stat)
                for el in soup:

                    # create a dictionary for each game
                    single[el.attrs['data-stat']] = el.text

                # add each single game to container of all games
                games.append(single)

        print(len(games)) # check if data is being scraped correctly

    return games
```
